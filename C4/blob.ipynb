```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Đọc hình ảnh đầu vào
image = cv2.imread('blob.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Khởi tạo đối tượng SimpleBlobDetector
params = cv2.SimpleBlobDetector_Params()
```

```python
# Cài đặt các thông số của SimpleBlobDetector
params.filterByArea = False
params.minArea = 1  # Diện tích tối thiểu của blob
params.minArea = 1000
params.filterByCircularity = False
params.filterByConvexity = False
params.filterByInertia = False
```

```python
# Khởi tạo detector
detector = cv2.SimpleBlobDetector_create(params)

# Dùng detector để phát hiện các blob
keypoints = detector.detect(image)
```

```python
# Vẽ các blob lên hình ảnh
image_with_keypoints = cv2.drawKeypoints(image, keypoints, np.array([]), (0, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Hiển thị hình ảnh với các blob đã phát hiện
cv2_imshow(image_with_keypoints)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
